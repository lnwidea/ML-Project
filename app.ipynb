```python
%%writefile app.py

import streamlit as st
import pandas as pd
import numpy as np
from pythainlp.corpus import thai_stopwords
from pythainlp.tokenize import word_tokenize
import joblib


model = joblib.load(open("/content/drive/MyDrive/ML Project/emotion_model.pkl", "rb"))

emotions = ["Sad", "Happy", "Love", "Angry", "Fear", "Surprise"]

thai_stopwords = list(thai_stopwords())

def text_process(text):
  final = "".join(u for u in text if u not in ("?", ".", ";", ":", "!", '"', "ๆ", "ฯ"))
  final = word_tokenize(final)
  final = " ".join(word for word in final)
  final = " ".join(word for word in final.split() if word.lower not in thai_stopwords)
  return final


def predict(text):
  x = model.predict([text])
  return x[0]


def percentage(text):
  x = model.predict_proba([text])
  list = []
  for i in x:
    i *= 100
  return x


def main():
  st.title("Thai Text Emotion Detection")
  st.subheader("Detect Emotions In Text (Only Thai text)")

  with st.form(key='my_form'):
    raw_text = st.text_area("Type Here")
    submit_text = st.form_submit_button(label='Submit')

  if submit_text:
    col = st.container()

    process_text = text_process(raw_text)
    prediction = predict(process_text)
    probability = percentage(process_text)

    with col:
      st.success("Prediction Results")
      st.write(emotions[prediction])
      st.success("Other probability results with percentage")
      df = pd.DataFrame(probability,columns=emotions)
      st.write(df)





if __name__ == '__main__':
  main()
```
